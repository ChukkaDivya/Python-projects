```python
def cubes(n):
    for i in range(1,n):
        yield i**3
    
```

```python
g=cubes(5)
```

```python
type(g)
```

```python
g
```

```python
for i in g:
    print(i)
```

```python
print(g)
```

```python
def gen_fib(n):
    a=0
    b=1
    for i in range(n):
        yield a
        a,b=b,a+b
```

```python
for i in gen_fib(2):
    print(i)
```

```python
g=gen_fib
```

```python
type(g)
```

```python
print(g)
```
