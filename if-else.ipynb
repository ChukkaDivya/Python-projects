Given a string, return a string where for every character in the original there are three characters

```python
def bust(a,b,c):
    if (a,b,c) in range(1,12):
        s=a+b+c
        if s<=21:
            return s
        elif s>21:
            if 11 in (a,b,c):
                return s-10
            else:
                return 'BUST'
```
