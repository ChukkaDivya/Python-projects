```python
import requests
import bs4
```

```python
first_url="https://quotes.toscrape.com/"
```

```python
res=requests.get(first_url)
```

```python
soup=bs4.BeautifulSoup(res.text,'lxml')
```

```python
#str(soup)
```

```python
soup.select(".author")
```

```python
authors=set()
for i in soup.select(".author"):
    authors.add(i.text)
```

```python
base_url="https://quotes.toscrape.com/page/{}/"
```

```python
all_author=set()
for i in range(1,11):
    page=base_url.format(i)
    result=requests.get(page)
    soup=bs4.BeautifulSoup(result.text,"lxml")
    authors=soup.select(".author")
    for author in authors:
        all_author.add(author.text)
        
```
