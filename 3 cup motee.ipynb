```python
from random import shuffle
```

```python
def lets_shuffle(list1):
    shuffle(list1)
    return list1
```

```python
def guess_here():
    guess=''
    while guess not in [0,1,2]:
        guess=int(input('pick a num from 0,1or 2'))
    return guess    
```

```python
def check_guess(list1,guess):
    if list1[guess]=='O':
        print('Yahoo correct')
    else:
        print('Try again !')
```

```python
list1=['','O','']
list2=lets_shuffle(list1)
guess=guess_here()
check_guess(list2,guess)
```
