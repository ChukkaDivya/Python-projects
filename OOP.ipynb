```python
class Circle():
    pi=3.14
    def __init__(self,radius=1):
        self.radius=radius
    def area(self):
        area=Circle.pi*(self.radius**2)
        print('area of circle is {} of radius is {}'.format(area,self.radius))
    def circumference(self):
        circumference=2*Circle.pi*self.radius
        print(f'circumference of circle is {circumference} ')
        
```

```python
c1=Circle(2)
```

```python
c1.circumference()
```

```python
c1.area()
```

```python
#Inheritance
class Animal():
    def __init__(self):
        print("ANIMAL CREATED !")
    def who_am_i(self):
        print('Iam a animal')
    def eat(self):
        print("Iam eating")
```

```python
class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("DOG CREATED !")
    def play(self):
        print("I love playing woof !")
    def who_am_i(self):
        print("Iam a dog")
        
        
```

```python
class Lion():
    def __init__(self,name):
        self.name=name
    def speak(self):
        return self.name + ' says arghh !'
```

```python
class Cat():
    def __init__(self,name):
        self.name=name
    def speak(self):
        return self.name + ' says meow !'
```

```python
leo=Lion("Leo")
```

```python
pam=Cat('Pam')
```

```python
leo.speak()
```

```python
pam.speak()
```

```python
for pet in [leo,pam]:
    print(type(pet))
    print(type(pet.speak()))
```

```python
def pet_speak(pet):
    print(pet.speak())
```

```python
pet_speak(leo)
```

```python
pet_speak(pam)
```

```python
class Book():
    def __init__(self,name,author,pages):
        self.name=name
        self.author=author
        self.pages=pages 
    def __str__(self):
        return f'the {self.name} by {self.author}'
    def __len__(self):
        return self.pages
```

```python
b=Book('strangers','joey william',456)
```

```python
print(b)
```
